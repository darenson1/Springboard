Once I took the original variables and created a new formula called Playmaker Rating, which provides a better weighted assessment of that given players all around personal stats, from both offensive and defesive standpoint. ### 2) Can you find some trends (high, low, increase, decrease, anomalies)?

There appears to be a good distribution of players who scored high and low with in the Playmaker and Scorer Rating Spectrum. When taking a closer look at some of the outliers, I noticed that many whom may have had a high rating may have resulted in the fact that they had only played in a very few amount of games, which would explain why their rating was so high. They would eventually have regressed towards the mean had they played in more games during the season which could have gone either positive or negatively for them. To solve that issue, I sorted players that to qualify for these rating purpose, they must have played in at least 55 games by the time the regular season (82 games) was over. Once I filtered for the qualified players, the distribution appeared more normal.
